```python
def drop_missing_percent(data,percentage):

  columns = data.columns
  percentage_missing = data.isna().sum()/ len(data) * 100

  missing_values_df = pd.DataFrame(
      {'column_name':columns,
       'percentage_missing':percentage_missing}
  )

  missing_drop = list(missing_values_df[missing_values_df.percentage_missing > percentage].column_name)

  data = data.drop(missing_drop,axis=1,inplace =True)

  return data


```
